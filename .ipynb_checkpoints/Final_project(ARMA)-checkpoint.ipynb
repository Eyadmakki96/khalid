```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
        
adf_test(df_train['Actual_Price'])
```

## Autoregressive Moving Average - ARMA(p,q)
In this first section we'll look at a stationary dataset, determine (p,q) orders.

```python
model_arma1_1= sm.tsa.arima.ARIMA(df_train.Actual_Price, order=(1,0,1))
results_arma1_1 = model_arma1_1.fit()
results_arma1_1.summary()
```

```python
start = len(df_train);
end = len(df_train) + len(df_test)-1
predictions = results_arma1_1.predict(start,end).rename('ARMA (1,1) Predictions')
title = 'Daily Electricity Prices'
ylabel=' Electricity Prices'
xlabel=''

ax = df_test['Actual_Price'].plot(legend=True,figsize=(12,6),title=title)
predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
model_arma_2_1= sm.tsa.arima.ARIMA(df_train.Actual_Price, order=(2,0,1))
results_arma_2_1= model_arma_2_1.fit()
results_arma_2_1.summary()
```

```python
model_arma_2_2= sm.tsa.arima.ARIMA(df_train.Actual_Price, order=(2,0,2))
results_arma_2_2= model_arma_2_2.fit()
results_arma_2_2.summary()
```

```python
predictions = results_arma_2_2.predict(start,end)
title = 'Daily Electricity Prices'
ylabel=' Electricity Prices'
xlabel=''

ax = df_test['Actual_Price'].plot(legend=True,figsize=(12,6),title=title)
predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```
